```python
import xarray as xr
from pathlib import Path
import os
import shutil
```

```python
dir_source = Path('/space/hall7/sitestore/eccc/crd/cccma/users/rpg002/data/intchl')

for model in [ 'CanESM5-CanOE_0', 'CanESM5-CanOE_1']:
    for dir in list((dir_source / 'assimilation' / model).glob('*nc')):

        name = dir.name
        outname = Path('/'.join(str(dir).split('/')[:-1])) / ('int' + name)
        ds = xr.open_dataset(dir)
        ds.where(ds.lev<150, drop = True).integrate('lev').to_netcdf(outname)
        os.remove(dir)
```

```python

for run in ['assimilation_bgc0', 'assimilation_bgc1', 'assimilation_bgc2', 'assimilation_bgc3']:
    for dir in list((dir_source / run / 'CanESM5').glob('*nc')):

        name = dir.name
        outname = Path('/'.join(str(dir).split('/')[:-1])) / ('int' + name)
        ds = xr.open_dataset(dir)
        ds.where(ds.lev<150, drop = True).integrate('lev').to_netcdf(outname)
        os.remove(dir)
```
