```python
import xarray as xr
import matplotlib.pyplot as plt
import pandas as pd
```

```python
rtd_dir = '/space/hall7/sitestore/eccc/crd/cccma/users/rpg002/data/rtd/'

var_list = ['TALz', 'TAL', 'DIC']
assimilation_CanOE = 'assimilation_CanOE_1'
```

```python



rtd_dict_carbon = {}
for var in var_list:
    rtd_dict_carbon[var] = {}
    for run in ['assimilation_CMOC', assimilation_CanOE, 'historical_CanOE', 'historical_CMOC']:
        if 'assim' in run:
            if 'CanOE' in run:
                dir_rtd = rtd_dir + 'assimilation/CanESM5-CanOE_1/'
            else:
                dir_rtd = rtd_dir + 'assimilation/CanESM5/'
        elif 'hist' in run:
            if 'CanOE' in run:
                dir_rtd = rtd_dir + 'historical/CanESM5-CanOE/'
            else:
                dir_rtd = rtd_dir + 'historical/CanESM5/'      
        rtd_dict_carbon[var][run] = xr.open_mfdataset(dir_rtd + '*carbon*')[var]


```

```python
var = 'TALz'
for run, ds in rtd_dict_carbon[var].items():
    ds['time'] = pd.to_datetime(ds['time'].astype(str))
    ds['time'] = [pd.to_datetime(str(t)) for t in ds['time'].values]


    ds.isel(depth = 0 ).plot(label = run)
    plt.legend()
```

```python
var = 'TAL'
plt.subplot(2,1,1)
rtd_dict_carbon[var]['assimilation_CMOC'].plot(label = 'Assimilation_CMOC')
plt.xlabel('')
# plt.xticks([])
plt.legend()
plt.subplot(2,1,2)
rtd_dict_carbon[var]['assimilation_CanOE_1'].plot(label = 'assimilation_CanOE_1', color = 'tab:orange')
plt.legend()
```

```python
var = 'TAL'
plt.subplot(2,1,1)
rtd_dict_carbon[var]['historical_CMOC'].plot(label = 'historical_CMOC')
plt.ticklabel_format(style='plain', axis='y')
plt.xlabel('')
# plt.xticks([])
plt.legend()
plt.subplot(2,1,2)
rtd_dict_carbon[var]['historical_CanOE'].plot(label = 'historical_CanOE', color = 'tab:orange')
plt.ticklabel_format(style='plain', axis='y')
plt.legend()
```

```python
var = 'DIC'
for run, ds in rtd_dict_carbon[var].items():
    ds['time'] = pd.to_datetime(ds['time'].astype(str))
    ds['time'] = [pd.to_datetime(str(t)) for t in ds['time'].values]


    ds.plot(label = run)
    plt.legend()
```
